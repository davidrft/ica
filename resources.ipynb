## Central Limit Theorem

```python
u = np.random.uniform(-1, 1, 1000000)

result = []

for _ in range(10000):
    idx = np.random.randint(0, 100000, 30)
    result.append(np.mean(u[idx]))
    
plt.hist(u)
plt.ylabel("Frequency")
plt.xlabel("Sample")
plt.savefig("figs/central_limit_theorem02.eps")
plt.show()
```

```python
plt.hist(result)
plt.ylabel("Frequency")
plt.xlabel("Average of samples")
plt.savefig("figs/central_limit_theorem01.eps")
plt.show()
```

## G Functions

```python
x = np.linspace(-3, 3, 2000)
y1 = g1(x)[0]
y2 = g2(x)[0]
y3 = g3(x)[0]

plt.plot(x, y1, label=r'$G(u) = \tanh(au)$')
plt.plot(x, y2, label=r'$G(u) = \frac{1}{a}\log(\cosh(au))$')
plt.plot(x, y3, label=r'$G(u) = -e^{u^{2}/2}$')
plt.legend(loc='upper center')
plt.ylabel(r"G(u)")
plt.xlabel(r"u")
plt.savefig("figs/gfunctions.eps")
plt.show()
```
